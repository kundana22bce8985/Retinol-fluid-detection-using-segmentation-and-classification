```python
import torch
from torch.utils.data import Dataset, DataLoader
import scipy.io # Import scipy.io
import numpy as np # Import numpy

class ChiuOCTDataset(Dataset):
    def __init__(self, data_path, transform=None):
        self.data_path = data_path
        self.transform = transform
        self.subject_files = [f for f in os.listdir(data_path) if f.endswith(".mat")]
        self.images = []
        self.masks = []

        for subject_file in self.subject_files:
            full_path = os.path.join(data_path, subject_file)
            data = scipy.io.loadmat(full_path)
            oct_vol = data["images"].astype(np.float32)
            mask_vol = np.nan_to_num(data["manualFluid1"]).astype(np.uint8)

            for i in range(oct_vol.shape[2]):
                self.images.append(oct_vol[:, :, i])
                self.masks.append(mask_vol[:, :, i])

        self.images = np.array(self.images)
        self.masks = np.array(self.masks)

    def __len__(self):
        return len(self.images)

    def __getitem__(self, idx):
        img = np.expand_dims(self.images[idx], axis=0)
        mask = np.expand_dims(self.masks[idx], axis=0)
        img = torch.tensor(img, dtype=torch.float32)
        mask = torch.tensor(mask, dtype=torch.float32)
        if self.transform:
            img, mask = self.transform(img, mask)
        return img, mask

dataset = ChiuOCTDataset(data_path)
dataloader = DataLoader(dataset, batch_size=4, shuffle=True)
```

```python
import torch.nn as nn

class DiceBCELoss(nn.Module):
    def __init__(self, smooth=1e-6):
        super().__init__()
        self.bce = nn.BCEWithLogitsLoss()
        self.smooth = smooth

    def forward(self, pred, target):
        bce_loss = self.bce(pred, target)
        pred = torch.sigmoid(pred)
        pred_flat = pred.view(-1)
        target_flat = target.view(-1)
        intersection = (pred_flat * target_flat).sum()
        dice_loss = 1 - (2 * intersection + self.smooth) / \
                    (pred_flat.sum() + target_flat.sum() + self.smooth)
        return bce_loss + dice_loss
```

```python
import torch
import torch.nn as nn

class DoubleConv(nn.Module):
    """(convolution => [BN] => ReLU) * 2"""
    def __init__(self, in_channels, out_channels, mid_channels=None):
        super().__init__()
        if not mid_channels:
            mid_channels = out_channels
        self.double_conv = nn.Sequential(
            nn.Conv2d(in_channels, mid_channels, kernel_size=3, padding=1, bias=False),
            nn.BatchNorm2d(mid_channels),
            nn.ReLU(inplace=True),
            nn.Conv2d(mid_channels, out_channels, kernel_size=3, padding=1, bias=False),
            nn.BatchNorm2d(out_channels),
            nn.ReLU(inplace=True)
        )

    def forward(self, x):
        return self.double_conv(x)

class Down(nn.Module):
    """Downscaling with maxpool then double conv"""
    def __init__(self, in_channels, out_channels):
        super().__init__()
        self.maxpool_conv = nn.Sequential(
            nn.MaxPool2d(2),
            DoubleConv(in_channels, out_channels)
        )

    def forward(self, x):
        return self.maxpool_conv(x)

class Up(nn.Module):
    """Upscaling then double conv"""
    def __init__(self, in_channels, out_channels, bilinear=True):
        super().__init__()

        # if bilinear, use the normal convolutions to reduce the number of channels
        if bilinear:
            self.up = nn.Upsample(scale_factor=2, mode='bilinear', align_corners=True)
            self.conv = DoubleConv(in_channels, out_channels, in_channels // 2)
        else:
            self.up = nn.ConvTranspose2d(in_channels, in_channels // 2, kernel_size=2, stride=2)
            self.conv = DoubleConv(in_channels, out_channels)

    def forward(self, x1, x2):
        x1 = self.up(x1)
        # input is CHW
        diffY = x2.size()[2] - x1.size()[2]
        diffX = x2.size()[3] - x1.size()[3]

        x1 = nn.functional.pad(x1, [diffX // 2, diffX - diffX // 2,
                                    diffY // 2, diffY - diffY // 2])
        # if you have padding issues, see
        # https://github.com/HaiyongJiang/U-Net-Pytorch-Unstructured-Buggy/commit/0e8545087dce92d633otoch.cat((x2, x1), dim=1)
        return self.conv(torch.cat([x2, x1], dim=1))


class OutConv(nn.Module):
    def __init__(self, in_channels, out_channels):
        super(OutConv, self).__init__()
        self.conv = nn.Conv2d(in_channels, out_channels, kernel_size=1)

    def forward(self, x):
        return self.conv(x)

class UNet(nn.Module):
    def __init__(self, n_channels=1, n_classes=1, bilinear=True):
        super(UNet, self).__init__()
        self.n_channels = n_channels
        self.n_classes = n_classes
        self.bilinear = bilinear

        self.inc = (DoubleConv(n_channels, 64))
        self.down1 = (Down(64, 128))
        self.down2 = (Down(128, 256))
        self.down3 = (Down(256, 512))
        factor = 2 if bilinear else 1
        self.down4 = (Down(512, 1024 // factor))
        self.up1 = (Up(1024, 512 // factor, bilinear))
        self.up2 = (Up(512, 256 // factor, bilinear))
        self.up3 = (Up(256, 128 // factor, bilinear))
        self.up4 = (Up(128, 64, bilinear))
        self.outc = (OutConv(64, n_classes))

    def forward(self, x):
        x1 = self.inc(x)
        x2 = self.down1(x1)
        x3 = self.down2(x2)
        x4 = self.down3(x3)
        x5 = self.down4(x4)
        x = self.up1(x5, x4)
        x = self.up2(x, x3)
        x = self.up3(x, x2)
        x = self.up4(x, x1)
        logits = self.outc(x)
        return logits
```

```python
import os
import scipy.io
import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader
import numpy as np

# Tversky Loss
class TverskyLoss(nn.Module):
    def __init__(self, alpha=0.7, beta=0.3, smooth=1e-6):
        super(TverskyLoss, self).__init__()
        self.alpha = alpha
        self.beta = beta
        self.smooth = smooth

    def forward(self, pred, target):
        pred = torch.sigmoid(pred)
        tp = (pred * target).sum()
        fp = ((1 - target) * pred).sum()
        fn = (target * (1 - pred)).sum()
        tversky = (tp + self.smooth) / (tp + self.alpha * fp + self.beta * fn + self.smooth)
        return 1 - tversky

# Dice score
def dice_score(pred, target, smooth=1e-6):
    pred = pred.reshape(-1)
    target = target.reshape(-1)
    intersection = (pred * target).sum()
    return (2. * intersection + smooth) / (pred.sum() + target.sum() + smooth)

# Custom dataset: only disease-positive slices
class DiseaseDataset(Dataset):
    def __init__(self, subject_files, data_path):
        self.samples = []
        for subject_file in subject_files:
            mat_path = os.path.join(data_path, subject_file)
            data = scipy.io.loadmat(mat_path)
            oct_vol = data['images'].astype(np.float32)
            gt_vol = np.nan_to_num(data['manualFluid1']).astype(np.uint8)

            for slice_idx in range(oct_vol.shape[2]):
                gt_mask = gt_vol[:, :, slice_idx]
                if gt_mask.sum() > 0:  # disease present
                    img = oct_vol[:, :, slice_idx]
                    img_norm = (img - img.min()) / (img.max() - img.min() + 1e-6)
                    self.samples.append((img_norm, gt_mask))

    def __len__(self):
        return len(self.samples)

    def __getitem__(self, idx):
        img, mask = self.samples[idx]
        img_tensor = torch.tensor(img[np.newaxis], dtype=torch.float32)
        mask_tensor = torch.tensor(mask[np.newaxis], dtype=torch.float32)
        return img_tensor, mask_tensor

# Load dataset
dataset = DiseaseDataset(subject_files, data_path)
loader = DataLoader(dataset, batch_size=4, shuffle=True)

# Model setup
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = UNet().to(device)
optimizer = torch.optim.Adam(model.parameters(), lr=1e-3)
criterion = TverskyLoss(alpha=0.7, beta=0.3)

# Training loop
num_epochs = 25
best_dice = 0.0

for epoch in range(num_epochs):
    model.train()
    train_loss = 0.0
    dice_scores = []

    for imgs, masks in loader:
        imgs, masks = imgs.to(device), masks.to(device)
        optimizer.zero_grad()
        outputs = model(imgs)
        loss = criterion(outputs, masks)
        loss.backward()
        optimizer.step()
        train_loss += loss.item()

        preds = (torch.sigmoid(outputs) > 0.3).float()
        for i in range(imgs.shape[0]):
            dice_scores.append(dice_score(preds[i], masks[i]).item())

    avg_loss = train_loss / len(loader)
    avg_dice = sum(dice_scores) / len(dice_scores)
    print(f"Epoch [{epoch+1}/{num_epochs}] - Loss: {avg_loss:.4f} - Dice: {avg_dice:.4f}")

    if avg_dice > best_dice:
        best_dice = avg_dice
        torch.save(model.state_dict(), 'best_unet_model.pth')
        print("âœ… Saved best model with Dice score:", best_dice)

print(f"\nðŸŽ¯ Training complete. Best Dice Score: {best_dice:.4f}")
```

```python
import scipy.io
import torch
import numpy as np
import matplotlib.pyplot as plt
import os

# Setup
prediction_threshold = 0.3
model.load_state_dict(torch.load('best_unet_model.pth', map_location=device))
model.eval()

# Optional: create folder to save images
os.makedirs("visuals", exist_ok=True)

# Counter to limit visuals
shown = 0
max_images = 2  # ðŸ‘ˆ show only 2 images total

# Loop through subjects
for subject_file in subject_files:
    mat_path = os.path.join(data_path, subject_file)
    data = scipy.io.loadmat(mat_path)

    oct_vol = data['images'].astype(np.float32)
    gt_vol = np.nan_to_num(data['manualFluid1']).astype(np.uint8)

    for slice_idx in range(oct_vol.shape[2]):
        img_np = oct_vol[:, :, slice_idx]
        gt_mask_np = gt_vol[:, :, slice_idx]

        # Skip slices with no disease
        if gt_mask_np.sum() == 0:
            continue

        # Normalize and prepare input
        img_norm = (img_np - img_np.min()) / (img_np.max() - img_np.min() + 1e-6)
        img_tensor = torch.tensor(img_norm[np.newaxis, np.newaxis], dtype=torch.float32).to(device)

        with torch.no_grad():
            output = model(img_tensor)
            prob = torch.sigmoid(output)
            pred_mask = (prob > prediction_threshold).float()

        pred_np = pred_mask[0, 0].cpu().numpy()
        prob_np = prob[0, 0].cpu().numpy()

        # Dice score
        intersection = (pred_np * gt_mask_np).sum()
        dice_score = (2 * intersection) / (pred_np.sum() + gt_mask_np.sum() + 1e-6)

        # Visualize
        plt.figure(figsize=(16, 4))
        plt.subplot(1, 4, 1)
        plt.title(f"OCT Image\n{subject_file}, Slice {slice_idx}")
        plt.imshow(img_norm, cmap='gray')
        plt.axis("off")

        plt.subplot(1, 4, 2)
        plt.title("Ground Truth Mask")
        plt.imshow(gt_mask_np, cmap='Reds')
        plt.axis("off")

        plt.subplot(1, 4, 3)
        plt.title(f"Predicted Mask\nThreshold: {prediction_threshold}")
        plt.imshow(pred_np, cmap='Blues')
        plt.axis("off")

        plt.subplot(1, 4, 4)
        plt.title("Probability Map")
        plt.imshow(prob_np, cmap='viridis')
        plt.axis("off")

        plt.suptitle(f"Dice Score: {dice_score:.4f}", fontsize=14)
        plt.tight_layout()
        plt.show()

        shown += 1
        if shown >= max_images:
            break  # stop after showing 2 images

    if shown >= max_images:
        break  # stop looping subjects once 2 images are visualized
```
